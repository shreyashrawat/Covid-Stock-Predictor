```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("time_test_dataset.xlsx")
```

```python
data = data.drop([0], axis = 0)
```

```python
data.columns = ["Stock Index", "10th","11th","12th","13th","14th","15th"]
```

```python
data.head()
```

```python
data.describe()
```

```python
data.index = data["Stock Index"]
data = data.drop(["Stock Index"], axis = 1)
```

## 1 Feature Prediction 

```python
# Creating dataset to feed into LSTM
X = data["10th"]

y = data["11th"]

X = pd.concat([X, data["11th"], data["12th"], data["13th"], data["14th"]])

y = pd.concat([y, data["12th"], data["13th"], data["14th"], data["15th"]])
```

```python
# Checking the nature of dataset
max(X), min(X), np.mean(X)
```

```python
max(y), min(y), np.mean(y)
```

```python
# Filling empty values with 
X.fillna(value = 1.0, inplace = True)
y.fillna(value = 1.0, inplace = True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
x_train = np.reshape(np.array(x_train), (-1,1))
x_test = np.reshape(np.array(x_test), (-1,1))

y_train = np.reshape(np.array(y_train), (-1,1))
y_test= np.reshape(np.array(y_test), (-1,1))
```

```python
X = np.reshape(np.array(X), (-1,1))
y = np.reshape(np.array(y), (-1,1))
```

```python
scaler = MinMaxScaler()
scaler_y = MinMaxScaler()
X = scaler.fit_transform(X)
y = scaler_y.fit_transform(y)
```

```python
X = X.reshape(X.shape[0],  X.shape[1],1)
```

```python
x_train.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(100,input_shape = (x_train.shape[1],1), return_sequences = True),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.LSTM(100, return_sequences = True),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.LSTM(50),
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(1, activation = 'relu')
    
])
```

```python
model.compile(loss = 'mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train, y_train, validation_data = (x_test, y_test),batch_size=64, epochs=5,verbose = True)
```

```python
model.evaluate(x_test, y_test)
```

```python
preds = model.predict(x_test)
```

```python
fig = plt.figure(figsize = (12,8))
plt.plot(preds.reshape(-1,1)[100:200], label = "Predicted")
plt.plot(y_test[100:200], label = "Actual")
plt.legend()
```

```python
mean_squared_error(y_test, preds.reshape(-1,1))
```

```python
s = data["15th"].fillna(value = 1.0)
testing_data = np.reshape(np.array(s), (-1,1))
testing_data = scaler.transform(testing_data)
testing_data = testing_data.reshape(testing_data.shape[0],  testing_data.shape[1],1)
```

```python
predictions = model.predict(testing_data)
```

```python
put_ratio_pred = scaler_y.inverse_transform(predictions)
```

```python
pd.DataFrame(put_ratio_pred).to_csv("Put_ratio.csv")
```
